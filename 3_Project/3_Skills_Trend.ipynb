## How are in-demand skills trending for Data Analysts?

#### Import Libraries, Load Data, Clean Data

```python
# Importing Libraries
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt 
import seaborn as sns

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

#### Filter for Data Analyst Jobs in the US

```python
df_DA_US = df[(df['job_title_short'] == 'Data Analyst') & (df['job_country'] == 'United States')].copy()
```

#### Create New Column with Job Posted Month Number

```python
df_DA_US['job_posted_month_no'] = df_DA_US['job_posted_date'].dt.month
```

#### Explode Job Skills for Analysis

```python
df_DA_US_explode = df_DA_US.explode('job_skills')
```

#### Create Pivot Table of Job Skill Counts by Month Sort by Total

```python
df_DA_US_pivot = df_DA_US_explode.pivot_table(index='job_posted_month_no', columns='job_skills', aggfunc='size', fill_value=0)

df_DA_US_pivot.loc['Total'] = df_DA_US_pivot.sum()

df_DA_US_pivot = df_DA_US_pivot[df_DA_US_pivot.loc['Total'].sort_values(ascending=False).index]

df_DA_US_pivot = df_DA_US_pivot.drop('Total')
```

#### Change Counts to Percentages
##### Format Month Column

```python
DA_totals = df_DA_US.groupby('job_posted_month_no').size()
```

```python
df_DA_US_percent = df_DA_US_pivot.div(DA_totals/100, axis=0)

#Change month number to month name
df_DA_US_percent = df_DA_US_percent.reset_index()
df_DA_US_percent['job_posted_month'] = df_DA_US_percent['job_posted_month_no'].apply(lambda x: pd.to_datetime(str(x), format='%m').strftime('%b'))
df_DA_US_percent = df_DA_US_percent.set_index('job_posted_month')
df_DA_US_percent = df_DA_US_percent.drop(columns='job_posted_month_no')

df_DA_US_percent
```

```python
df_plot = df_DA_US_percent.iloc[:, :5]

sns.lineplot(data=df_plot, dashes=False, palette='tab10') 
sns.set_theme(style='ticks')
sns.despine()

plt.title('Trending Top Skills for Data Analysts in the US')
plt.xlabel('2023')
plt.ylabel('Likelihood in Job Posting')
plt.legend().remove()

from matplotlib.ticker import PercentFormatter
ax = plt.gca()
ax.yaxis.set_major_formatter(PercentFormatter(decimals=0))

plt.tight_layout()
plt.annotate(
        df_plot.columns[0],  # Text to display
        xy=(11, df_plot.iloc[-1, 0]),  # Position of the point to annotate
        xytext=(11.5, df_plot.iloc[-1, 0] + 0.5 * 0),  # Adjusted text position
        arrowprops=dict(arrowstyle='-', color='gray')  # Optional arrow to show connection
    )

plt.annotate(
        df_plot.columns[1],  # Text to display
        xy=(11, df_plot.iloc[-1, 1]),  # Position of the point to annotate
        xytext=(11.5, df_plot.iloc[-1, 1] + 0.5 * 1),  # Adjusted text position
        arrowprops=dict(arrowstyle='-', color='gray')  # Optional arrow to show connection
    )
plt.annotate(
        df_plot.columns[2],  # Text to display
        xy=(10, df_plot.iloc[-1, 2]),  # Position of the point to annotate
        xytext=(11.5, df_plot.iloc[-1, 2] + 1.5 * 2),  # Adjusted text position
        arrowprops=dict(arrowstyle='-', color='gray', shrinkB=15)  # Optional arrow to show connection
    )
plt.annotate(
        df_plot.columns[3],  # Text to display
        xy=(10, df_plot.iloc[-1, 3]),  # Position of the point to annotate
        xytext=(11.5, df_plot.iloc[-1, 3] - 1.5 * 3),  # Adjusted text position
        arrowprops=dict(arrowstyle='-', color='gray', shrinkB=30)  # Optional arrow to show connection
    )
plt.annotate(
        df_plot.columns[4],  # Text to display
        xy=(11, df_plot.iloc[-1, 4]),  # Position of the point to annotate
        xytext=(11.5, df_plot.iloc[-1, 4] - 0.5 * 4),  # Adjusted text position
        arrowprops=dict(arrowstyle='-', color='gray')  # Optional arrow to show connection
    )
plt.show()
```
