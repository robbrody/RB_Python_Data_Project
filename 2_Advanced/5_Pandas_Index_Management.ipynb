```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt 

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])

df.set_index('job_posted_date', inplace=True)

df.reset_index(drop=True, inplace=True)

df.head()
```

```python
df.pivot_table(values='salary_year_avg', index='company_name', aggfunc='median').sort_index(ascending=False).head(5)
```

```python
job_salary_median = df.pivot_table(values='salary_year_avg', index='job_title_short', aggfunc='median')

job_salary_median.reset_index(inplace=True)

job_salary_median.set_index('salary_year_avg', inplace=True)

job_salary_median.head()
```
